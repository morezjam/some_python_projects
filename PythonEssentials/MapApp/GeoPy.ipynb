```python
import pandas

df = pandas.read_csv("10.3 supermarkets-semi-colons.txt", delimiter = ";")
df = df.set_index(["ID"])

df["Address"] = df.Address + ", " + df.City + ", " + df.State + ", " + df.Country
```

```python
from geopy.geocoders import ArcGIS

arc = ArcGIS()

df["Coordinates"] = df["Address"].apply(lambda x: arc.geocode(x))
df
```

```python
df["Y"] = df["Coordinates"].apply(lambda x: x.longitude)
df["X"] = df["Coordinates"].apply(lambda x: x.latitude)

df
```

```python
xy = df["X"].apply(lambda x: str(x)) + ", " + df["Y"].apply(lambda y: str(y))
xy = xy.apply(lambda l: arc.reverse(l))
list(xy)
```
